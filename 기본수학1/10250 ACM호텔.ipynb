# 10250 ACM호텔

```python
t = int(input()) #테스트케이스 수


for i in range(t):
    cnt =1
    h,w,n = map(int,input().split())  
    #h=호텔의 층 수, w=각 층의 방수 n=몇번째 손님인지
    while True:
        if n > h : 
            n= n-h
            cnt += 1
        else:
            break
 
    print(n*100 + cnt)
```

* 테스트 케이스 T만큼 반복한다.
* h=호텔의 층 수, w=각 층의 방수 n=몇번째 손님인지
* while문을 반복하며, 손님의 순서가 호텔의 층수보다 크다면, 호텔의 층수만큼 빼주는 동시에, 
* cnt를 1씩 증가시켜, 즉 몇번째 호실에 묵을지를 결정해준다.
* 손님의 순서가 층 수보다 작아지게 되면 while문을 탈출하고,
* n은 층 수이므로 정수 100을 곱하고, cnt는 호수를 결정짓는 변수이므로 (n*100 + cnt)를 출력한다.
