# 2292 벌집

```python
n = int(input()) 
i =1 
while n>1 :
    n -= (6 * i ) #방 = (n//6*i)
    i+= 1
    
print(i)
```

### 육각형으로 이루어진 벌집이 있다. 중앙의 방 1부터 시작해서 이웃하는 방에 돌아가면서 1씩 증가하는 번호를 주소로 매길 수 있다. 숫자 N이 주어졌을 때, 벌집의 중앙 1에서 N번 방까지 최소 개수의 방을 지나서 갈 때 몇 개의 방을 지나가는지를 계산하는 프로그램이다.

* 벌집은 방이 늘어날 때마다 6의 배수로 늘어난다.
* 예를 들어 2-7까지는 방이 2이고, 8-19까지는 방이 3개이다.

* while문을 사용하여 n이 1 이상일 때, 6의 배수만큼 감소시키고, 방의 수(i)는 1씩 증가시킨다.
* n이 1보다 작아지면while문을 탈출하고, 방의 개수를 출력한다.
